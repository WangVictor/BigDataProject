```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
data = pd.read_csv('./NYPD_Complaint_Data_Historic.csv')
```

```python
ass_data = data[data['OFNS_DESC'].str.contains('FELONY',na=False)]
rob_data = data[data['OFNS_DESC'].str.contains('ROBBERY',na=False)]
lar_data = data[data['OFNS_DESC'].str.contains('GRAND LARCENY',na=False)]
mur_data = data[data['OFNS_DESC'].str.contains('MURDER',na=False)]
moto_data = data[data['OFNS_DESC'].str.contains('MOTOR',na=False)]
rape_data = data[data['OFNS_DESC'].str.contains('SEX CRIMES',na=False)]
burg_data = data[data['OFNS_DESC'].str.contains('BURGLARY',na=False)]
```

```python
yearly_complains = []
for year in range(2005,2016):
    yearly_complains.append(data['CMPLNT_FR_DT'].str.contains(str(year)).sum())
yearly_rape_complains = []
for year in range(2005,2016):
    yearly_rape_complains.append(rape_data['CMPLNT_FR_DT'].str.contains(str(year)).sum())
yearly_burg_complains = []
for year in range(2005,2016):
    yearly_burg_complains.append(burg_data['CMPLNT_FR_DT'].str.contains(str(year)).sum())
yearly_ass_complains = []
for year in range(2005,2016):
    yearly_ass_complains.append(ass_data['CMPLNT_FR_DT'].str.contains(str(year)).sum())
yearly_lar_complains = []
for year in range(2005,2016):
    yearly_lar_complains.append(lar_data['CMPLNT_FR_DT'].str.contains(str(year)).sum())
yearly_mur_complains = []
for year in range(2005,2016):
    yearly_mur_complains.append(mur_data['CMPLNT_FR_DT'].str.contains(str(year)).sum())
yearly_moto_complains = []
for year in range(2005,2016):
    yearly_moto_complains.append(moto_data['CMPLNT_FR_DT'].str.contains(str(year)).sum())
yearly_rob_complains = []
for year in range(2005,2016):
    yearly_rob_complains.append(rob_data['CMPLNT_FR_DT'].str.contains(str(year)).sum())
```

```python
%matplotlib inline
# These are the "Tableau 20" colors as RGB.    
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),    
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]    
  
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.    
for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.)       
```

```python
plt.figure(figsize=(8, 6))    
  
# Remove the plot frame lines. They are unnecessary chartjunk.    
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)    
  
# Ensure that the axis ticks only show up on the bottom and left of the plot.    
# Ticks on the right and top of the plot are generally unnecessary chartjunk.    
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()    
  
# Limit the range of the plot to only where the data is.    
# Avoid unnecessary whitespace.    
plt.ylim(0, 62000)    
plt.xlim(2006, 2016)    
  
# Make sure your axis ticks are large enough to be easily read.    
# You don't want your viewers squinting to read your plot.    
#plt.yticks(range(0, 91, 10), [str(x) + "%" for x in range(0, 91, 10)], fontsize=14)    
plt.xticks(range(2006,2016),fontsize=10)    
  
# Provide tick lines across the plot to help your viewers trace along    
# the axis ticks. Make sure that the lines are light and small so they    
# don't obscure the primary data lines.    
for y in range(0, 62000, 10000):    
    plt.plot(range(2006, 2016), [y] * len(range(2006, 2016)), "--", lw=0.5, color="black", alpha=0.3)    

plt.tick_params(axis="both", which="both", bottom="off", top="off",    
                labelbottom="on", left="off", right="off", labelleft="on")    
  
#plt.plot(range(2005,2016),yearly_complains,lw=2.5, color=tableau20[1])  
#plt.text(2015.5, 468576, 'Total Crimes', fontsize=14, color=tableau20[1])  
plt.plot(range(2006,2016),yearly_rob_complains[1:],lw=2.5,color =tableau20[2])
plt.text(2015.5, 16913, 'Robbery', fontsize=14, color=tableau20[2])  
plt.plot(range(2006,2016),yearly_rape_complains[1:],lw=2.5,color =tableau20[3])
plt.text(2015.5, 1500, 'Rape', fontsize=14, color=tableau20[3]) 
plt.plot(range(2006,2016),yearly_burg_complains[1:],lw=2.5,color =tableau20[4])
plt.text(2015.5, 15000, 'Burglary', fontsize=14, color=tableau20[4]) 
plt.plot(range(2006,2016),yearly_ass_complains[1:],lw=2.5,color =tableau20[5])
plt.text(2015.5, 20685, 'Felony Assault', fontsize=14, color=tableau20[5]) 
plt.plot(range(2006,2016),yearly_lar_complains[1:] ,lw=2.5,color =tableau20[6])
plt.text(2015.5, 49315, 'Grand Lareny', fontsize=14, color=tableau20[6]) 
plt.plot(range(2006,2016),yearly_mur_complains[1:] ,lw=2.5,color =tableau20[7])
plt.text(2015.5, 0, 'Murder', fontsize=14, color=tableau20[7]) 
plt.plot(range(2006,2016),yearly_moto_complains[1:] ,lw=2.5,color =tableau20[8])
plt.text(2015.5, 7319, 'Grand Lareny of Motor Vehile', fontsize=14, color=tableau20[8]) 
#plt.text(1995, 93, "Percentage of Bachelor's degrees conferred to women in the U.S.A."    
#       ", by major (1970-2012)", fontsize=17, ha="center")    
  
plt.text(2006.0, -10000, "Data source: https://data.cityofnewyork.us/Public-Safety/NYPD-Complaint-Data-Historic"    
       "\nAuthor: Police Department (NYPD)" , fontsize=10)    
  
#plt.savefig("percent-bachelors-degrees-women-usa.png", bbox_inches="tight") 
```
